```python
import calendar
import random
class Caregiver:
    #intialize
    def __init__(self, name: str, phone: str, email: str, rate: float, hrs: int):
        self.name = name
        self.phone = phone
        self.email = email
        self.rate = rate
        self.hrs = hrs
        self.avail = {}

    #set caregivers availability
    def randomize_avail(self, days_in_month: int):
        valid_availability = ["available", "unavailable", "preferred"]
        for day in range(1, days_in_month + 1):
            # Randomly assign "AM" and "PM" availability for each day
            self.avail[day] = {
                "AM": random.choice(valid_availability),
                "PM": random.choice(valid_availability),
            }

    @staticmethod
    def generate_schedule(caregivers, month, year):
        #get days in given month/year
        days_in_month = calendar.monthrange(year, month)[1]
        schedule = {}
        #intialize schedule with empty cells for shifts
        for day in range (1, days_in_month + 1):
            schedule[day] = {'AM': [], 'PM': []}

        #Iterate through each caregiver and day to add them to schedule
        for caregiver in caregivers:
            for day in range(1, days_in_month + 1):
                for shift in ['AM', 'PM']:
                    if caregiver.avail.get(day, {}).get(shift) in ['preferred', 'available']:
                        schedule[day][shift].append(caregiver.name)
        
        return schedule

    @staticmethod
    #Calculate weekly pay for each caregiver
    def calculate_weekly_pay(caregivers):
        weekly_pay_report = {}
        for caregiver in caregivers:
            weekly_hours = caregiver.hrs
            weekly_pay = weekly_hours * caregiver.rate
            weekly_pay_report[caregiver.name] = {
                'weekly_hours': weekly_hours,
                'weekly_pay': weekly_pay
            }
        return weekly_pay_report

    @staticmethod
    #Print schedule for given month/year
    def print_schedule(schedule, month, year):
        print(f"Care Schedule for {calendar.month_name[month]} {year}")
        days_in_month = calendar.monthrange(year, month)[1]

        for day in range(1, days_in_month + 1):
            print(f"\nDay {day}")
            print("AM Shift: ", ", ".join(schedule[day]['AM']) if schedule [day]['AM'] else "No caregiver assigned")
            print("PM Shift: ", ", ".join(schedule[day]['PM']) if schedule [day]['PM'] else "No caregiver assigned")
    
    @staticmethod
    #Generate html report for caregivers pay/schedule
    def html_report(caregivers, schedule, month, year):
        weekly_pay_report = Caregiver.calculate_weekly_pay(caregivers)
        total_weekly = sum(c['weekly_pay'] for c in weekly_pay_report.values())
        total_monthly = total_weekly * 4

        #Start of HTML doc
        html = f"""
        <html>
        <head>
            <title>Caregiver Weekly Pay Report</title>
        </head>
        <body>
            <h1>Weekly Pay Report for {calendar.month_name[month]} {year}</h1>
            <table border = "1">
                <tr>
                    <th>Caregiver Name</th>
                    <th>Weekly Hours</th>
                    <th>Weekly Pay</th>
                </tr>
        """

        #Add caregiver's pay details to table in HTML
        for caregiver in caregivers:
            caregiver_name = caregiver.name
            weekly_hours = weekly_pay_report[caregiver_name]['weekly_hours']
            weekly_pay = weekly_pay_report[caregiver_name]['weekly_pay']
            html += f"""
                <tr>
                    <td>{caregiver_name}</td>
                    <td>{weekly_hours}</td>
                    <td>${weekly_pay:.2f}</td>
                </tr>
            """

    #Add total weekly pay to html
        html += f"""
            <tr>
                <td><strong>Total Weekly Pay</strong></td>
                <td></td>
                <td><strong>${total_weekly:.2f}</strong></td>
            </tr>
            <tr>
            </table>
            
            <h1>Caregiver Schedule</h1>
            <h2>Visual Calendar for {calendar.month_name[month]} {year}</h2>
            <table border = "1">
                <tr>
                    <th>Sun</th>
                    <th>Mon</th>
                    <th>Tue</th>
                    <th>Wed</th>
                    <th>Thu</th>
                    <th>Fri</th>
                    <th>Sat</th>
                </tr>
        """

        #Add caregiver schedule for month into table
        month_calendar = calendar.monthcalendar(year, month)
        for week in month_calendar:
            html += "<tr>"
            for day in week:
                if day == 0:
                    html += "<td></td>" #leaves cells blank that are not within scope of the month
                else:
                    am_shift = ", ".join(schedule[day]['AM']) if schedule[day]['AM'] else "No caregiver assigned"
                    pm_shift = ", ".join(schedule[day]['PM']) if schedule[day]['PM'] else "No caregiver assigned"
                    html += f"""
                <td>
                    <strong>{day}</strong><br>
                    AM: {am_shift}<br>
                    PM: {pm_shift}
                </td>
                """    
            html += "</tr>"

        #close HTML tags
        html += """
            </table>
            </body>
            </html>
        """

        #save html content to file
        with open(f"caregiver_pay_report_{month}_{year}.html", "w") as file:
            file.write(html)
        #print file name of report
        print(f"HTML report: caregiver_pay_report_{month}_{year}.html")

        return html

#Main
if __name__ == "__main__":
    #Sample list of caregiver data
    caregivers = [
        Caregiver("Alex", "111-1111", "Alex@umd.edu", 20, 12),
        Caregiver("Edward", "222-1111", "Edward@umd.edu", 20, 36),
        Caregiver("Chassen", "333-1111", "Chassen@umd.edu", 20, 18),
        Caregiver("Angelina", "444-1111", "Angelina@umd.edu", 20, 42),
        Caregiver("Dude", "555-1111", "Dude@umd.edu", 20, 48),
        Caregiver("Karen", "666-1111", "Karen@umd.edu", 20, 30),
        Caregiver("Sharon", "777-1111", "Sharon@umd.edu", 20, 24),
        Caregiver("Mary", "888-1111", "Mary@umd.edu", 20, 48),
    ]
    
    #Randomly assign availability
    month, year = 11, 2024
    days_in_month = calendar.monthrange(year,month)[1]
    for caregiver in caregivers:
        caregiver.randomize_avail(days_in_month)

    #Generate schedule
    schedule = Caregiver.generate_schedule(caregivers, month, year)

    #Print schedule
    Caregiver.print_schedule(schedule, month, year)
    
    #Generate HTML schedule
    Caregiver.html_report(caregivers, schedule, month, year)
```
